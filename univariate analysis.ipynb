```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/main/day20-univariate-analysis/train.csv')
data.head()
```

# 1.categorical data

a.count plot

```python
sns.countplot(data=data,x='Survived')
```

```python
data['Embarked'].value_counts().plot(kind='bar',color=['red','green','blue'])
```

b. Pie chart

```python
plt.pie(data['Sex'].value_counts(),autopct='%.2f',labels=['Male','Female'])
plt.plot
```

```python
data['Survived'].value_counts().plot(kind='pie',autopct='%.2f')
```
