```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('https://raw.githubusercontent.com/lucko515/regression-python/master/Salaries%20based%20on%20Positions%20-%20dataset/Position_Salaries.csv')
data
```

```python
#split to dependent and independent variebles
x=data.iloc[:,1:2]
y=data.iloc[:,2:]
```

```python
#building linear regression model
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x,y)
```

```python
#building polynomail regression model
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2)
x_poly=poly.fit_transform(x)
lr2=LinearRegression()
lr2.fit(x_poly,y)
```

```python
x_poly
```

```python
#visualize the result for linear regression
plt.scatter(x,y,c='Blue')
plt.plot(x,lr.predict(x),c='Red')
plt.title("Bluff detection model(Linear Regression)")  
plt.xlabel("Position Levels")  
plt.ylabel("Salary")  
plt.show() 
```

```python
#visualize the result for polynomial regression
plt.scatter(x,y,c='Blue')
plt.plot(x,lr2.predict(x_poly),c='Red')
plt.title("Bluff detection model(polynomial Regression)")  
plt.xlabel("Position Levels")  
plt.ylabel("Salary")  
plt.show() 
```

```python
#for degree=3
#visualize the result for polynomial regression
poly3=PolynomialFeatures(degree=3)
x_poly=poly3.fit_transform(x)
lr3=LinearRegression()
lr3.fit(x_poly,y)
```

```python
plt.scatter(x,y,c='Blue')
plt.plot(x,lr3.predict(x_poly),c='Red')
plt.title("Bluff detection model(polynomial Regression)")  
plt.xlabel("Position Levels")  
plt.ylabel("Salary")  
plt.show() 
```

```python
#for degree=4
#visualize the result for polynomial regression
poly4=PolynomialFeatures(degree=4)
x_poly=poly4.fit_transform(x)
lr4=LinearRegression()
lr4.fit(x_poly,y)
```

```python
plt.scatter(x,y,c='Blue')
plt.plot(x,lr4.predict(x_poly),c='Red')
plt.title("Bluff detection model(polynomial Regression)")  
plt.xlabel("Position Levels")  
plt.ylabel("Salary")  
plt.show() 
```

```python
#predicting the final result with linear regression
lin_predict=lr.predict([[6.5]])
lin_predict
```

```python
#predicting the final result with polynomial regression
poly_predict=lr4.predict(poly4.fit_transform([[6.5]]))
poly_predict
```
