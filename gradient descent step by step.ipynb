```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x,y=make_regression(n_samples=6,
    n_features=1,
    n_informative=1,
    n_targets=1,
    noise=90,
    random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
#apply linear regression
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(x,y)
```

```python
lin_reg.coef_
```

```python
lin_reg.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,lin_reg.predict(x),color='Red')
```

```python
# Lets apply Gradient Descent assuming slope is constant m = 47.41
# and let's assume the starting value for intercept b = 0
y_pred=((47.41*x)+100).reshape(6)
y_pred
```

```python
plt.scatter(x,y)
plt.plot(x,lin_reg.predict(x),color='Red',label='Lin_Reg')
plt.plot(x,y_pred,color='Green',label='b=0')
plt.legend()
plt.show()
```

```python
#iteration 1
m = 47.41
b = 100

loss_slope = -2 * np.sum(y - m*x.ravel() - b)
loss_slope
```

```python
#lets take learning rate is 0.1
ler_rate=0.1
step_size=loss_slope*ler_rate
step_size
```

```python
#calculating the new intercept
b=b-step_size
b
```

```python
y_pred1 = ((47.41 * x) + b).reshape(6)

plt.scatter(x,y)
plt.plot(x,lin_reg.predict(x),color='red',label='Lin_Reg')
plt.plot(x,y_pred1,color='#00a65a',label='b = {}'.format(b))
plt.plot(x,y_pred,color='Green',label='b = 0')
plt.legend()
plt.show()
```

```python
#iteration 2
loss_slope = -2 * np.sum(y - m*x.ravel() - b)
loss_slope
```

```python
step_size=loss_slope*ler_rate
step_size
```

```python
#calculating the new intercept
b=b-step_size
b
```

```python
y_pred2 = ((47.41 * x) + b).reshape(6)

plt.scatter(x,y)
plt.plot(x,lin_reg.predict(x),color='red',label='Lin_Reg')
plt.plot(x,y_pred1,color='#00a65a',label='b = {}'.format(b))
plt.plot(x,y_pred2,color='Blue',label='b = {}'.format(b))
plt.plot(x,y_pred,color='Green',label='b = 0')
plt.legend()
plt.show()
```

```python
#iteration 3
loss_slope = -2 * np.sum(y - m*x.ravel() - b)
loss_slope
```

```python
step_size=loss_slope*ler_rate
step_size
```

```python
#calculating the new intercept
b=b-step_size
b
```

```python
y_pred3 = ((47.41 * x) + b).reshape(6)

plt.scatter(x,y)
plt.plot(x,lin_reg.predict(x),color='red',label='Lin_Reg')
plt.plot(x,y_pred1,color='#00a65a',label='b = {}'.format(b))
plt.plot(x,y_pred2,color='Blue',label='b = {}'.format(b))
plt.plot(x,y_pred3,color='Yellow',label='b = {}'.format(b))
plt.plot(x,y_pred,color='Green',label='b = 0')
plt.legend()
plt.show()
```

```python
b = 100
m = 47.41
lr = 0.01

epochs = 100

for i in range(epochs):
    loss_slope = -2 * np.sum(y - m*x.ravel() - b)
    b = b - (lr * loss_slope)

    y_pred = m * x + b

    plt.plot(x,y_pred)

plt.scatter(x,y)
```
