```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/krishnaik06/Multiple-Linear-Regression/master/50_Startups.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#separating dependent and independent variables
x=df.iloc[:,:-1]
y=df.iloc[:,-1:]
```

```python
x.shape,y.shape
```

```python
x
```

```python
y
```

```python
states=pd.get_dummies(x['State'],dtype='float')
X=x.drop('State',axis=1)
X=pd.concat([X,states],axis=1)
X.head()
```

```python
#splitting the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
#fitting to our MLR model to the training set
from sklearn.linear_model import LinearRegression
mlr=LinearRegression()
mlr.fit(x_train,y_train)
```

```python
y_pred=mlr.predict(x_test)
x_pred=mlr.predict(x_train)
```

```python
y_pred
```

```python
y_test
```

```python
mlr.score(x_train,y_train)
```

```python
mlr.score(x_test,y_test)
```

```python
from sklearn.metrics import r2_score #using r2 score
print('Testing accuracy: ',r2_score(y_test,y_pred))
print('Training accuracy: ',r2_score(y_train,x_pred))
```
